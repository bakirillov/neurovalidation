### Model

```python
def get_convolutional_model(latent_space_size, activation="relu", optimizer="adam", loss="mae", bias=True):
    inp = Input(shape=(10,), name="TIPI")
    rshp = Reshape(target_shape=(1,2,5))(inp)
    conv = Conv2D(5, (1,2), strides=1, use_bias=bias, padding="valid", activation="elu")(rshp)
    rshp2 = Reshape(target_shape=(5,))(conv)
    model = Model(inputs=inp, outputs=rshp2)
    model.compile(loss=loss, optimizer=optimizer)
    return(model)
```

```python
class PermutationCallback(Callback):
    
    def __init__(self, filename, train, test, **kwargs):
        super(self.__class__, self).__init__(**kwargs)
        self.filename = filename
        self.test = test
        self.train = train
        self.matrices = []
        self.weights = []
    
    @staticmethod
    def regression_report(real, pred):
        r = np.zeros((3,5))
        funs = [mean_absolute_error, mean_squared_error]
        for f in range(2):
            for a in range(5):
                r[f,a] = funs[f](real[:,a], pred[:,a])
        return(r)
    
    def on_train_begin(self, logs={}):
        if not op.exists(op.split(self.filename)[0]):
            os.makedirs(op.split(self.filename)[0])
        
    def on_epoch_end(self, epoch, logs={}):
        self.matrices.append(
            PermutationCallback.regression_report(
                self.test, self.model.predict(self.train)
            )
        )
        self.weights.append(
            self.model.layers[2].get_weights()[0]
        )
        
    def on_train_end(self, logs={}):
        np.save(
            self.filename,
            np.stack(self.matrices)
        )
        np.save(
            self.filename+"_weights",
            np.stack(self.weights)
        )
```

```python
get_model = get_convolutional_model
```

### Training a lot of models

```python
def mean_history(model_f, model_p, data, n_train, n_epochs, figsize):
    H_DIR = "../models/nn/histories"
    if not op.exists(H_DIR):
        os.makedirs(H_DIR)
    for a in range(n_train):
        print("Model "+str(a))
        train_X, test_X, train_Y, test_Y = train_test_split(
            data[data.columns[1:11]].values, data[pfq].values, test_size=0.4
        )
        cb = PermutationCallback(op.join(H_DIR, str(a)+".path.npy"), test_X, test_Y)
        model = model_f(**model_p)
        history = model.fit(
            train_X, train_Y, validation_data=(test_X, test_Y), 
            epochs=n_epochs, shuffle=True, verbose=0,
            callbacks=[cb]
        )
        np.save(op.join(H_DIR, str(a)+".loss.npy"), history.history["loss"])
        np.save(op.join(H_DIR, str(a)+".val_loss.npy"), history.history["val_loss"])
        K.clear_session()
```

```python
with timeit():
    mean_history(get_model, {"latent_space_size": 8, "loss": "mse", "bias":False}, data, 100, 500, (10,10))
```

```python
def plot_losses_with_error(losses, val_losses, title, bayes):
    mean_loss = np.mean(losses, axis=0)
    max_loss = np.max(losses, axis=0)
    min_loss = np.min(losses, axis=0)
    mean_v_loss = np.mean(val_losses, axis=0)
    max_v_loss = np.max(val_losses, axis=0)
    min_v_loss = np.min(val_losses, axis=0)
    plt.subplot(2,1,1)
    plt.title(title)
    plt.plot(list(range(mean_loss.shape[0])), mean_loss, color="black")
    plt.fill_between(list(range(mean_loss.shape[0])), min_loss, max_loss, color="lightgrey")
    plt.grid(True)
    plt.xlabel("#epochs")
    plt.ylabel("Train loss")
    plt.axhline(y=bayes, color="grey", linestyle='-.')
    plt.subplot(2,1,2)
    plt.plot(list(range(mean_v_loss.shape[0])), mean_v_loss, color="black")
    plt.fill_between(list(range(mean_v_loss.shape[0])), min_v_loss, max_v_loss, color="lightgrey")
    plt.grid(True)
    plt.xlabel("#epochs")
    plt.ylabel("Validation loss")
    plt.axhline(y=bayes, color='grey', linestyle='-.')
    plt.show()
```

### Permutations

```python
def permutation_test(model_f, model_p, data, n_perm, n_epochs):
    print("Real data")
    REAL_DIR = "../models/nn/real"
    PERM_DIR = "../models/nn/permutations"
    pfq = ["E.5PFQ", "A.5PFQ", "C.5PFQ", "ES.5PFQ", "O.5PFQ"]
    for a in range(n_perm):
        train_X, test_X, train_Y, test_Y = train_test_split(
            data[data.columns[1:11]].values, data[pfq].ix[np.random.permutation(data.shape[0])].values, 
            test_size=0.4
        )
        print("Permutation "+str(a))
        C_PERM_DIR = op.join(PERM_DIR, str(a))
        p_train_ind = np.random.choice(list(range(train_Y.shape[0])), size=train_Y.shape[0])
        p_train = train_Y[p_train_ind]
        p_test_ind = np.random.choice(list(range(test_Y.shape[0])), size=test_Y.shape[0])
        p_test = test_Y[p_test_ind]
        cb = PermutationCallback(op.join(C_PERM_DIR, str(a)+".path.npy"), test_X, test_Y)
        model = model_f(**model_p)
        history = model.fit(
            train_X, p_train, validation_data=(test_X, p_test), 
            epochs=n_epochs, shuffle=True, verbose=0,
            callbacks=[cb]
        )
        np.save(op.join(C_PERM_DIR, str(a)+".loss.npy"), history.history["loss"])
        np.save(op.join(C_PERM_DIR, str(a)+".val_loss.npy"), history.history["val_loss"])
        K.clear_session()
```

```python
with timeit():
    permutation_test(get_model, {"latent_space_size": 8, "loss": "mae", "bias": False}, data, 100, 500)
```

```python
def plot_permutations(perm, real, figsize, out_dir, fps=24, animate=False):
    if not op.exists(out_dir):
        os.makedirs(out_dir)
    filenames = []
    ksses = []
    fig = plt.figure(figsize=figsize)
    G = gridspec.GridSpec(2, 5)
    funs = ["MAE", "MSE"]
    chars = ["E", "A", "C", "ES", "O"]
    for b in range(perm.shape[1]):
        current_perm = perm[:,b,:,:]
        current_real = real[:,b,:,:]
        out = op.join(out_dir, str(b)+".png")
        filenames.append(out)
        ks = []
        for a in range(2):
            for b in range(5):
                axes = fig.add_subplot(G[a, b])
                plt.title(funs[a]+" for "+chars[b])
                plt.hist(current_perm[:,a,b], color="lightgrey")
                plt.hist(current_real[:,a,b], color="grey")
                ks.append(
                    ks_2samp(current_perm[:,a,b], current_real[:,a,b])[1]
                )
        fig.savefig(out)
        fig.clf()
        del current_perm
        del current_real
        del axes
        gc.collect(2)
        print(out)
        ksses.append(ks)
    np.save(op.join(out_dir, "ks.npy"), np.stack(ksses))
    if animate:
        fnames = sorted(filenames, key = lambda x: int(op.split(x)[-1].split(".")[0]))
        clip = ImageSequenceClip(fnames, fps=fps)
        clip.write_gif(op.join(out_dir, "animation.gif")) 
```

```python
with timeit():
    plot_permutations(perm_paths, real_paths, (10,10), "../models/nn/comparison/", animate=True, fps=24)
```

```python
def animate_weights(weights, filepath, animate=True, figsize=(10,10), fps=24):
    if not op.exists(filepath):
        os.makedirs(filepath)
    fig = plt.figure(figsize=figsize)
    filenames=[]
    for b in range(weights.shape[0]):
        i = 1
        out = op.join(filepath, str(b) + ".png")
        for a in range(5):
            axes = fig.add_subplot(5,2,i)
            plt.imshow(weights[b,:,:,a,:].reshape(2,5), cmap="Greys")
            i += 1
            plt.subplot(5,2,i)
            plt.imshow(weights[b,:,:,:,a].reshape(2,5), cmap="Greys")
            i += 1
        fig.savefig(out)
        filenames.append(out)
        print(out)
        fig.clf()
        del axes
        gc.collect(2)
    if animate:
        fnames = sorted(filenames, key = lambda x: int(op.split(x)[-1].split(".")[0]))
        clip = ImageSequenceClip(fnames, fps=fps)
        clip.write_gif(op.join(filepath, "animation.gif")) 
```
